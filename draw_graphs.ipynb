```python
import numpy as np
bins = ['Original', 'Gaussian\nNoise', 'Defocus\nBlur', 'Elastic\nTransform', 'Motion\nBlur\n(H)',
        'Motion\nBlur\n(V)', 'Zoom\nBlur', "Shot", 'Impulse', 'Contrast']

baseline_acc = np.array([86.63, 60.48, 85.28, 82.72, 67.37, 65.59, 85.87, 73.03, 45.05, 66.37])
baseline_err = 100 - baseline_acc
joint_acc = np.array([87.07, 63.21, 86.07, 83.01, 66.25, 66.10, 85.97, 74.89, 49.38, 70.23])
joint_err = 100 - joint_acc
ttt_acc = np.array([87.05, 63.24, 86.09, 83.10, 66.43, 66.76, 86.03, 74.60, 49.81, 70.95])
ttt_err = 100 - ttt_acc
ttto_acc = np.array([86.74, 64.54, 86.56, 83.77, 68.00, 67.42, 86.25, 75.87, 51.35, 71.43])
ttto_err = 100 - ttto_acc
```

```python
import matplotlib.pyplot as plt

x = np.arange(len(bins))

fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111)
width = 0.2
ax.bar(x, baseline_err, width, label='Baseline ResNet18')
ax.bar(x + width, joint_err, width, label='Jointly Trained')
ax.bar(x + 2 * width, ttt_err, width, label='TTT')
ax.bar(x + 3 * width, ttto_err, width, label='TTT-Online')
ax.set_title("Test Errors on Corrupted GTSRB")
ax.set_ylabel('Error Rate %')
ax.set_xticks(x + 1.5 * width)
ax.set_xticklabels(bins)
#for tick in ax.get_xticklabels():
#   tick.set_rotation(30)
ax.legend(loc='best')
fig.savefig('results.png')
```
